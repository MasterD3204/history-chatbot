```python
def get_random_seed_chunk():
    """
    Lấy ngẫu nhiên một object từ Weaviate để làm 'hạt giống'.
    Phiên bản này sử dụng cú pháp đã được xác nhận là đúng: include_vector=True.
    """
    try:
        agg_response = collection.aggregate.over_all(total_count=True)
        total_count = agg_response.total_count

        if total_count == 0:
            print("Lỗi: Collection không có object nào.")
            return None

        # Thử tối đa 5 lần để tìm một lô có chứa object hợp lệ
        for _ in range(5):
            batch_size = 10
            if total_count <= batch_size:
                random_offset = 0
            else:
                random_offset = random.randint(0, total_count - batch_size)
     
            response = collection.query.fetch_objects(
                limit=batch_size,
                offset=random_offset,
                include_vector=True
            )

            if response.objects:
                random.shuffle(response.objects)

            for seed_object in response.objects:
                vector_data = seed_object.vector.get('default') if seed_object.vector else None

                if vector_data:  # Nếu tìm thấy vector
                    return {
                        "id": str(seed_object.uuid),
                        "content": seed_object.properties.get('content', ''),
                        "vector": vector_data
                    }

            print("Cảnh báo: Không tìm thấy object nào có vector trong lô hiện tại. Thử lại...")

        print("Lỗi nghiêm trọng: Không thể tìm thấy object có vector sau nhiều lần thử.")
        return None

    except Exception as e:
        print(f"Không thể lấy seed chunk ngẫu nhiên: {e}")
        return None

def find_similar_chunks(seed_vector, limit=20):
    """
    Tìm kiếm các chunk tương đồng dựa trên vector hạt giống (cú pháp v4).
    Hàm này đã đúng từ trước nhưng giữ lại để đảm bảo tính nhất quán.
    """
    try:
        response = collection.query.near_vector(
            near_vector=seed_vector,
            limit=limit,
            return_properties=["content"],
            return_metadata=wvc.query.MetadataQuery(distance=True)
        )

        similar_chunks = []
        for item in response.objects:
            similar_chunks.append({
                "id": str(item.uuid),
                "content": item.properties.get('content', '')
            })
        return similar_chunks
    except Exception as e:
        print(f"Lỗi khi tìm kiếm tương đồng: {e}")
        return []

def rerank_chunks(seed_chunk, candidates):
    """Xếp hạng lại các chunk ứng viên (không thay đổi)."""
    query_content = seed_chunk['content']
    pairs = [[query_content, candidate['content']] for candidate in candidates]

    scores = reranker_model.predict(pairs)

    for i, candidate in enumerate(candidates):
        candidate['rerank_score'] = scores[i]

    candidates.sort(key=lambda x: x['rerank_score'], reverse=True)
    return candidates

def generate_question_with_gemini(context_chunks):
    """Tạo câu hỏi và câu trả lời từ các chunk ngữ cảnh (không thay đổi)."""
    prompt_template = """Bạn là một chuyên gia trong việc tạo ra các câu hỏi phức tạp để đánh giá hệ thống tìm kiếm thông tin.
Dựa trên kiến thức tổng hợp từ 5 đoạn văn bản được cung cấp dưới đây, hãy thực hiện các yêu cầu sau:

1.  **Phân tích và Tổng hợp:** Đọc và hiểu sâu sắc thông tin từ TẤT CẢ 5 đoạn văn để nắm bắt các mối liên kết giữa chúng.
2.  **Tạo Câu hỏi Tự nhiên:** Soạn ra MỘT câu hỏi duy nhất mà câu trả lời đòi hỏi phải kết hợp thông tin từ ÍT NHẤT 3 trong số 5 đoạn văn.
    - **YÊU CẦU BẮT BUỘC:** Câu hỏi phải hoàn toàn tự nhiên, giống như một người dùng thông thường đang tìm kiếm thông tin, không biết trước về các đoạn văn bản này.
    - **TUYỆT ĐỐI KHÔNG** được sử dụng các cụm từ như "dựa trên văn bản", "theo thông tin đã cho", "trong các đoạn trích sau", v.v. Câu hỏi phải đứng độc lập.
3.  **Tạo Câu trả lời:** Cung cấp câu trả lời chi tiết cho câu hỏi bạn vừa tạo, chỉ dựa vào thông tin có trong các đoạn văn.

**Các đoạn văn:**
{context_string}

**Định dạng kết quả của bạn:**

**Câu hỏi:**
[Viết câu hỏi tự nhiên tại đây]

**Câu trả lời:**
[Viết câu trả lời chi tiết tại đây]
"""
    context_string = ""
    for i, chunk in enumerate(context_chunks):
        context_string += f"**Đoạn {i+1} (ID: {chunk['id']}):**\n\"{chunk['content']}\"\n\n"

    prompt = prompt_template.format(context_string=context_string.strip())

    try:
        response = gemini_model.generate_content(prompt)
        return response.text
    except Exception as e:
        print(f"Lỗi khi gọi API Gemini: {e}")
        return None

def parse_gemini_response(response_text):
    """Phân tích văn bản trả về từ Gemini để lấy câu hỏi và câu trả lời (không thay đổi)."""
    question_match = re.search(r"\*\*Câu hỏi:\*\*\s*(.*?)\s*\*\*Câu trả lời:\*\*", response_text, re.DOTALL)
    answer_match = re.search(r"\*\*Câu trả lời:\*\*\s*(.*)", response_text, re.DOTALL)

    question = question_match.group(1).strip() if question_match else None
    answer = answer_match.group(1).strip() if answer_match else None

    return question, answer

def main(num_samples_to_generate, output_file="generated_dataset.jsonl"):
    """Hàm chính để chạy toàn bộ quy trình."""

    with open(output_file, 'w', encoding='utf-8') as f:
        for _ in tqdm(range(num_samples_to_generate), desc="Đang tạo dataset"):
            # 1. Lấy seed chunk
            seed_chunk = get_random_seed_chunk()
            if not seed_chunk or 'vector' not in seed_chunk or seed_chunk['vector'] is None:
                print("Bỏ qua: Không lấy được seed chunk hợp lệ hoặc không có vector.")
                continue

            # 2. Tìm kiếm tương đồng
            similar_chunks = find_similar_chunks(seed_chunk['vector'], limit=10)
            similar_chunks = [c for c in similar_chunks if c['id'] != seed_chunk['id']]
            if len(similar_chunks) < 4:
                print(f"Bỏ qua: Chỉ tìm thấy {len(similar_chunks)} chunk tương đồng.")
                continue

            # 3. Rerank
            reranked_chunks = rerank_chunks(seed_chunk, similar_chunks)

            # 4. Chọn top 4 chunk và kết hợp với seed chunk
            top_4_reranked = reranked_chunks[:4]
            final_ground_truths = [seed_chunk] + top_4_reranked

            # 5. Tạo câu hỏi với Gemini
            gemini_response = generate_question_with_gemini(final_ground_truths)
            if not gemini_response:
                print("Bỏ qua: Gemini không trả về kết quả.")
                continue

            # 6. Parse kết quả
            question, answer = parse_gemini_response(gemini_response)
            if not question or not answer:
                print("Bỏ qua: Không thể parse câu hỏi/câu trả lời từ Gemini.")
                print("--- Response từ Gemini ---")
                print(gemini_response)
                print("-------------------------")
                continue

            # 7. Lưu kết quả
            dataset_entry = {
                "question": question,
                "ground_truth_ids": [chunk['id'] for chunk in final_ground_truths],
                "generated_answer": answer,
                "ground_truth_contents": [chunk['content'] for chunk in final_ground_truths]
            }

            f.write(json.dumps(dataset_entry, ensure_ascii=False) + '\n')

    print(f"\nHoàn thành! Dataset đã được lưu vào file '{output_file}'.")

if __name__ == "__main__":
    NUM_SAMPLES = 100 
    main(num_samples_to_generate=NUM_SAMPLES)
```

```python
import jsonlines
import numpy as np
import pandas as pd
from tqdm import tqdm
from sentence_transformers import SentenceTransformer

# ==============================================================================
# PHẦN 1: IMPORT HỆ THỐNG TRUY XUẤT CỦA BẠN
# (Sao chép toàn bộ code retriever của bạn vào đây)
# ==============================================================================
# --- BẮT ĐẦU CODE TRUY XUẤT CỦA BẠN ---
import sys
import os
from collections import defaultdict
from typing import List, Dict

# Giả sử client và class_name đã được khởi tạo ở đâu đó
# Bạn cần đảm bảo các import này hoạt động trong môi trường của bạn


TOP_K_BM25 = 50
TOP_K_VEC  = 50
TOP_K_FINAL = 5 # Đây là số lượng tài liệu retriever của bạn trả về
RRF_K = 60

collection = client.collections.get("HistoryDocument")

def safe_extract_obj(o):
    obj_id = getattr(o, "uuid", None)
    props = getattr(o, "properties", {})
    content = props.get("content")
    score = None
    meta = getattr(o, "metadata", None)
    if meta:
        score = getattr(meta, "score", None) or getattr(meta, "distance", None)
    return {"uuid": str(obj_id), "content": content, "score": score}

def query_hybrid_alpha(query_text, vector, alpha, limit) -> List[Dict]:
    resp = collection.query.hybrid(
        query=query_text,
        vector=vector,
        alpha=alpha,
        limit=limit,
        query_properties=['content']
    )
    results = []
    for idx, o in enumerate(resp.objects):
        info = safe_extract_obj(o)
        info["rank"] = idx + 1
        results.append(info)
    return results

def rrf_fusion(lists_of_results, k=RRF_K, top_k=5):
    accum = defaultdict(float)
    meta = {}
    for res_list in lists_of_results:
        for item in res_list:
            docid = item["uuid"]
            rank = item.get("rank")
            if rank is None: continue
            accum[docid] += 1.0 / (k + rank)
            if docid not in meta:
                meta[docid] = {"content": item.get("content"), "source_ranks": []}
            meta[docid]["source_ranks"].append(rank)

    fused = [{"id": docid, "rrf_score": score, "content": meta[docid]["content"], "ranks": meta[docid]["source_ranks"]} for docid, score in accum.items()]
    fused.sort(key=lambda x: x["rrf_score"], reverse=True)
    return fused[:top_k]

def retriever_fn(query, embedding_model):
    query_vector = embedding_model.encode(query).tolist()
    bm25_results = query_hybrid_alpha(query, query_vector, alpha=0.0, limit=TOP_K_BM25)
    vec_results  = query_hybrid_alpha(query, query_vector, alpha=1.0, limit=TOP_K_VEC)
    top_final = rrf_fusion([bm25_results, vec_results], k=RRF_K, top_k=TOP_K_FINAL)
    return top_final
# --- KẾT THÚC CODE TRUY XUẤT CỦA BẠN ---


# ==============================================================================
# PHẦN 2: CÁC HÀM TÍNH TOÁN CHỈ SỐ
# ==============================================================================

def calculate_metrics(retrieved_ids: List[str], ground_truth_ids: List[str], k: int) -> Dict[str, float]:
    """Tính toán tất cả các chỉ số cho một giá trị k."""
    metrics = {}
    retrieved_at_k = retrieved_ids[:k]
    ground_truth_set = set(ground_truth_ids)

    # --- Các chỉ số cơ bản ---
    hits_at_k = sum(1 for doc_id in retrieved_at_k if doc_id in ground_truth_set)

    metrics['precision@k'] = hits_at_k / k if k > 0 else 0.0
    metrics['recall@k'] = hits_at_k / len(ground_truth_set) if len(ground_truth_set) > 0 else 0.0

    p = metrics['precision@k']
    r = metrics['recall@k']
    metrics['f1@k'] = 2 * (p * r) / (p + r) if (p + r) > 0 else 0.0

    metrics['hit_rate@k'] = 1.0 if hits_at_k > 0 else 0.0

    # --- MRR@k ---
    reciprocal_rank = 0.0
    for i, doc_id in enumerate(retrieved_at_k):
        if doc_id in ground_truth_set:
            reciprocal_rank = 1.0 / (i + 1)
            break
    metrics['mrr@k'] = reciprocal_rank

    # --- MAP@k ---
    average_precision = 0.0
    hits = 0
    if len(ground_truth_set) > 0:
        for i, doc_id in enumerate(retrieved_at_k):
            if doc_id in ground_truth_set:
                hits += 1
                average_precision += hits / (i + 1)
        metrics['map@k'] = average_precision / len(ground_truth_set)
    else:
        metrics['map@k'] = 0.0

    # --- NDCG@k ---
    relevance = [1 if doc_id in ground_truth_set else 0 for doc_id in retrieved_at_k]
    dcg = sum(relevance[i] / np.log2(i + 2) for i in range(k))

    ideal_relevance = sorted(relevance, reverse=True)
    idcg = sum(ideal_relevance[i] / np.log2(i + 2) for i in range(k))

    metrics['ndcg@k'] = dcg / idcg if idcg > 0 else 0.0

    return metrics

def evaluate(dataset_path: str, embedding_model, k_values: List[int]):
    """Chạy đánh giá trên toàn bộ dataset cho nhiều giá trị k."""

    # Khởi tạo cấu trúc để lưu trữ kết quả
    results = {k: defaultdict(float) for k in k_values}
    total_samples = 0

    with jsonlines.open(dataset_path) as reader:
        for sample in tqdm(reader, desc="Đang đánh giá"):
            question = sample['question']
            ground_truth_ids = sample['ground_truth_ids']

            retrieved_docs = retriever_fn(question, embedding_model)
            retrieved_ids = [doc['id'] for doc in retrieved_docs]

            for k in k_values:
                # Tính toán các chỉ số cho giá trị k hiện tại
                metrics_at_k = calculate_metrics(retrieved_ids, ground_truth_ids, k)
                for key, value in metrics_at_k.items():
                    results[k][key] += value # Cộng dồn để tính trung bình sau

            total_samples += 1

    # Tính toán trung bình và tạo DataFrame
    avg_results = []
    k_retrieved = TOP_K_FINAL # Số lượng tài liệu retriever trả về

    for k in k_values:
        avg_row = {f"{key.replace('@k','')}@{k}": value / total_samples for key, value in results[k].items()}
        avg_results.append(avg_row)

    df = pd.DataFrame(avg_results)
    df = df.set_index(pd.Index([f"k={k}" for k in k_values])).T # Chuyển vị để dễ đọc

    # Thêm các chỉ số Context
    context_precision = results[k_retrieved]['precision@k'] / total_samples
    context_recall = results[k_retrieved]['recall@k'] / total_samples

    print("\n" + "="*40)
    print("      KẾT QUẢ ĐÁNH GIÁ HỆ THỐNG TRUY XUẤT")
    print("="*40)
    print(f"Tổng số mẫu: {total_samples}")
    print(f"Số lượng tài liệu truy xuất (Context Size): {k_retrieved}")
    print("-" * 40)
    print(f"Context Precision: {context_precision:.4f}")
    print(f"Context Recall:    {context_recall:.4f}")
    print("-" * 40)
    print("CHI TIẾT THEO TỪNG K:")
    print(df.to_string(float_format="%.4f"))
    print("="*40 + "\n")


if __name__ == "__main__":
    DATASET_FILE = "generated_dataset.jsonl"
    K_VALUES_TO_EVALUATE = [1, 3, 5] 



    evaluate(DATASET_FILE, model, K_VALUES_TO_EVALUATE)
```
